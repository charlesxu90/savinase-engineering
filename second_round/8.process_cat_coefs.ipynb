# Draw Kcat/Km barplot

```python
import pandas as pd

df_cat = pd.read_excel('data/round2/wet-lab/final-kinetic_results.xlsx', sheet_name='Sheet2')
df_cat.head()
```

```python
df_cat_agg = df_cat.groupby('SampleID').agg(
    Vmax_mean=('Vmax(nm*L^-1/s)', 'mean'),
    Vmax_std=('Vmax(nm*L^-1/s)', 'std'),
    Km_mean=('Km(mM)', 'mean'),
    Km_std=('Km(mM)', 'std'),
    kcat_mean=('Kcat(s^-1)', 'mean'),
    kcat_std=('Kcat(s^-1)', 'std'),
    kcatKm_mean=('Kcat/Km(s^-1/M)', 'mean'),
    kcatKm_std=('Kcat/Km(s^-1/M)', 'std'),
)

df_cat_agg = df_cat_agg.reset_index()
df_cat_agg.to_csv('data/round2/wet-lab/top_mut_cat_coeffs_agg.csv', index=False)
df_cat_agg
```

# Prepare data to draw a latex table

```python
df_data = df_cat_agg.copy()
df_data['kcatKm_mean_2'] =  df_data['kcatKm_mean'] / 1e6  # scale to s^-1/uM
df_data['kcatKm_std_2'] = df_data['kcatKm_std'] / 1e6  # scale to s^-1/uM

df_data
```

```python
df_data.to_csv('data/round2/wet-lab/top_mut_cat_coeffs_agg_processed.csv', index=False)
```
