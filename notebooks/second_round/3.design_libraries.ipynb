# Merge the six libraries into one

```python
def create_pos_list(libraries):
    all_pos_list = [lib['pos'].tolist() for lib in libraries]
    uniq_pos = list(set([pos for pos_list in all_pos_list for pos in pos_list]))
    return sorted(uniq_pos)


lib_uniq_pos = create_pos_list([df_run1_lib, df_run2_lib, df_run3_lib, df_run4_lib, df_run5_lib, df_run6_lib])
print(len(lib_uniq_pos))
print(lib_uniq_pos)
```

```python
def create_pos2aa_dict(libraries):
    uniq_pos = create_pos_list(libraries)

    pos2aa = {pos: [] for pos in uniq_pos}

    for pos in uniq_pos:
        for df in libraries:
            if pos not in df['pos'].values:
                continue

            if df[df['pos'] == pos]['fix_aa'].values[0]:
                pos2aa[pos].append(df[df['pos'] == pos]['fix_aa'].values[0])

            elif pos in df['pos'].values:
                pos2aa[pos].extend(df[df['pos'] == pos]['sel_mut_aa'].values[0])

        pos2aa[pos] = list(set(pos2aa[pos]))

    return pos2aa

lib_pos2aa = create_pos2aa_dict([df_run1_lib, df_run2_lib, df_run3_lib, df_run4_lib, df_run5_lib, df_run6_lib])
print(lib_pos2aa)
```

```python
def create_pos2aa_count_dict(libraries, topn=1000):
    lib_pos2aa = create_pos2aa_dict(libraries)
    lib_pos2aa_count = {pos: {aa: 0 for aa in lib_pos2aa[pos]} for pos in lib_pos2aa}

    for pos in lib_pos2aa_count.keys():
        for aa in lib_pos2aa_count[pos].keys():
            for df in libraries:
                if pos not in df.pos.values:
                    continue

                if df[df.pos == pos]['fix_aa'].values[0] == aa:
                    lib_pos2aa_count[pos][aa] += topn

                elif aa in df[df.pos == pos]['sel_mut_aa'].values[0]:

                    if aa == df[df.pos == pos]['ref_aa'].values[0]:
                        lib_pos2aa_count[pos][aa] += df[df.pos == pos]['ref_count'].values[0]
                    else:
                        # get the idx of the aa in the mut_aa list
                        idx = df[df.pos == pos]['sel_mut_aa'].values[0].index(aa)
                        lib_pos2aa_count[pos][aa] += df[df.pos == pos]['mut_aa_count'].values[0][idx]
    
    return lib_pos2aa_count

lib_pos2aa_count = create_pos2aa_count_dict([df_run1_lib, df_run2_lib, df_run3_lib, df_run4_lib, df_run5_lib, df_run6_lib], topn=1000)
print(lib_pos2aa_count)
```

```python
import pandas as pd

def create_merged_lib_df(libraries, wt, lib2merge=6, topn=1000):
    lib_pos2aa_count = create_pos2aa_count_dict(libraries, topn=topn)
    df_lib_merged = pd.DataFrame({'pos': list(lib_pos2aa_count.keys()), 
                        'mut_aa': [list(lib_pos2aa_count[pos].keys()) for pos in lib_pos2aa_count.keys()],
                        'mut_aa_count': [list(lib_pos2aa_count[pos].values()) for pos in lib_pos2aa_count.keys()]
                        })
    df_lib_merged['ref_aa'] = [wt[pos-1] for pos in df_lib_merged.pos]
    df_lib_merged['ref_count'] = [lib2merge*topn - sum(row['mut_aa_count']) 
                            if row['ref_aa'] not in row['mut_aa'] 
                            else row['mut_aa_count'][row['mut_aa'].index(row['ref_aa'])] 
                            for i, row in df_lib_merged.iterrows()]

    df_lib_merged['sel_mut_aa'] = [row['mut_aa'] + [row['ref_aa']] 
                                    if row['ref_count'] > 0 and row['ref_aa'] not in row['mut_aa'] 
                                    else row['mut_aa'] 
                                    for i, row in df_lib_merged.iterrows()]
    # drop a pos if mut_aa is only wt_aa
    df_lib_merged['is_only_wt_aa'] = [True if len(row['sel_mut_aa']) == 1 and row['sel_mut_aa'][0] == row['ref_aa'] else False for i, row in df_lib_merged.iterrows()]
    df_lib_merged = df_lib_merged[~df_lib_merged['is_only_wt_aa']]
    df_lib_merged.drop(columns=['is_only_wt_aa'], inplace=True)

    # redefine fix_aa
    df_lib_merged['fix_aa'] = [row['mut_aa'][0] if row['mut_aa_count'][0] == lib2merge*topn else '' 
                                for i, row in df_lib_merged.iterrows()]

    return df_lib_merged

df_top_lib_merged = create_merged_lib_df([df_run1_lib, df_run2_lib, df_run3_lib, df_run4_lib, df_run5_lib, df_run6_lib], wt=savinase, lib2merge=6, topn=1000)
df_top_lib_merged
```

```python
# evaluate library size
from dgc_libgen import get_codons_from_aa
import numpy as np

df_top_lib_merged['dg_codons'], df_top_lib_merged['expanded_codons'] = zip(*df_top_lib_merged.apply(get_codons_from_aa, axis=1))
df_top_lib_merged['n_codons'] = [len(codons) for codons in df_top_lib_merged.dg_codons]
df_top_lib_merged['n_expanded_codons'] = [len(codons) for codons in df_top_lib_merged.expanded_codons]
library_size = np.prod([ n for n in df_top_lib_merged.n_expanded_codons if n > 0])
print(f"Library size: {library_size}")
```

```python
import numpy as np

def convert_into_bpn_position(positions):
    """ Convert from savinase positions to BPN positions """
    uniq_pos = list(set(positions))
    uniq_pos_sorted = np.sort(uniq_pos).tolist()

    # Convert the positions into savinase positions
    savinase = 'AQSVPWGISRVQAPAAHNRGLTGSGVKVAVLDTGI-STHPDLNIRGGASFVPGEPST-QDGNGHGTHVAGTIAALNNSIGVLGVAPSAELYAVKVLGASGSGSVSSIAQGLEWAGNNGMHVANLSLGSPSPSATLEQAVNSATSRGVLVVAASGNSG-AGS---ISYPARYANAMAVGATDQNNNRASFSQYGAGLDIVAPGVNVQSTYPGSTYASLNGTSMATPHVAGAAALVKQKNPSWSNVQIRNHLKNTATSLGSTNLYGSGLVNAEAATR'

    bpn_pos = []
    idx_diff = 0
    for i, aa in enumerate(savinase):
        if aa == '-': idx_diff += 1

        if i+1 in uniq_pos_sorted:
            bpn_pos.append(i+1 + idx_diff)
    
    # restore the original order
    pos_mapping = dict(zip(uniq_pos_sorted, bpn_pos))
    bpn_pos = [pos_mapping[p] for p in positions]
    return bpn_pos

convert_into_bpn_position([57])
```
